```python
#adding in what I need to make this work
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import cross_val_score
from sklearn.metrics import auc_score
```

```python
#importing the csv for test
test = pd.read_csv('/Users/jessicagarson/learning-python/gadsdata/lemons/test.csv')
```

```python
#import the csv for train
train = pd.read_csv('/Users/jessicagarson/learning-python/gadsdata/lemons/train.csv')
```

```python
#lets take a look at what this looks like 
train.head()
```

```python
#checking out the test dataframe
test.head()
```

```python
#making text words into vector
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(train.Make)
X_test = vectorizer.transform(test.Model)
```

```python
#cross validation on make by age
cross_val_score(LogisticRegression(), X_train, train.VehicleAge)
```

```python
#Fitting the model
model = LogisticRegression().fit(X_train, list(train.VehicleAge))
```

```python
#making predictions
predictions = model.predict_proba(X_test)[:,1]
```

```python
#creating a solution dataframe
solution = pd.DataFrame({'id': test.RefId, 'bad_buy': predictions})
solution.to_csv('lemon_solution.csv', index=False)
```
