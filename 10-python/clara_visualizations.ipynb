New job, new project!

TMF's core business is its newsletter services, which give stock recommendations to subscribers. At any given time, each service may have well over 100 recommended stocks. And, since TMF is all about value investment and true ownership of companies, the collection of recommended stocks for a given service includes many companies that are considered good ones to own, but whose stocks are not at a great buying price right now. The purpose of the project is to determine whether these "Buy Now" recommendations actually outperform the collection of recommended stocks as a whole. That is to say, if an investor were to buy only the "Buy Now" stocks, would her returns be better than if she chose from the collection of recommendations at large?

## Build All-Recommendation Porfolio
In order to determine how the "buy now" picks compare to the recommendations as a whole, we need a way to model the average investment returns.

This is perhaps the most difficult part of the project, as it requires domain knowledge in an area where I have very little. I want to be able to build the portfolios in such a way that a small number of high-growth stocks won't dominate the returns, which will require some periodic rebalance. That seems complicated, so, to begin, I'm just going to simply (and naively) invest 10 currency units every day, divided equally between recommendations.

```python
own_shares_bool = prices_df.applymap(lambda x: False if x <=0 else True)
```

```python
from decimal import Decimal as D
```

```python
daily_investment = D(10)
```

```python
number_recs = own_shares_bool.apply(lambda x: D(sum(x)), axis=1)
```

```python
plt.plot(number_recs.index, number_recs, lw=2.5)
plt.xlabel('date')
plt.ylabel('number of stocks')
plt.title('number of recommended stocks in service')
plt.show()
```

```python
shares_per_currency_unit = prices_df.applymap(lambda x: D(0) if x <=0 else D(1)/x)
```

```python
shares_to_purchase = shares_per_currency_unit.apply(lambda x: x / number_recs * daily_investment, axis=0)
```

```python
portfolio_shares = shares_to_purchase.apply(cumsum, axis=0) * own_shares_bool
```

```python
portfolio_values = portfolio_shares * prices_df
```

The cumsum function would have wiped out my "sold" indicator, so I multiplied it by a boolean data frame that records whether or not my portfolio owns each share, to zero out stocks after they have been sold.

```python
shares1 = portfolio_shares['960902396']
```

```python
plt.plot(shares1.index, shares1, lw=2.5)
plt.xlabel('date')
plt.ylabel('shares')
plt.title('shares owned of ticker ' + shares1.name + ' over time')
plt.show()
```

Cool, that looks like it worked as expected. This stock was picked up early in the history of the service, so the rate of purchase was high in the beginning when there were few other stocks in the service. You can see that the rate of share purchase increased when the price fell, and ownership does indeed fall to zero when expected.

Now, let's try plotting the value of the portfolio overall.

```python
total_value = portfolio_values.apply(sum, axis=1)
```

```python
plt.plot(total_value.index, total_value, lw=2.5)
plt.xlabel('date')
plt.ylabel('currency')
plt.title('all-recommendation portfolio value')
plt.show()
```
